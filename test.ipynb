Dưới đây là tổng hợp các cú pháp (syntax) thường được sử dụng trong biểu thức chính quy (regex) trong Python:

1. Ký tự Thường
Mô tả: Khớp chính xác với ký tự đó trong chuỗi.
Ví dụ: abc sẽ khớp với chuỗi "abc".
2. Dấu Chấm .
Mô tả: Khớp với bất kỳ ký tự đơn nào ngoại trừ ký tự xuống dòng (\n).
Ví dụ: a.c sẽ khớp với "abc", "a1c", nhưng không khớp với "ac".
3. Dấu Gạch Chéo \
Mô tả: Dùng để thoát các ký tự đặc biệt hoặc biểu diễn các ký tự đặc biệt.
Ví dụ: \. sẽ khớp với dấu chấm thực sự, \d sẽ khớp với bất kỳ chữ số nào (0-9).
4. Mốc Bắt Đầu ^ và Kết Thúc $
Mô tả:
^: Khớp với vị trí bắt đầu của chuỗi.
$: Khớp với vị trí kết thúc của chuỗi.
Ví dụ: ^hello sẽ khớp với "hello world" nhưng không khớp với "say hello".
5. Ký Tự Trong Dấu Ngoặc Vuông [ ]
Mô tả: Khớp với bất kỳ một ký tự nào trong ngoặc vuông.
Ví dụ: [aeiou] sẽ khớp với bất kỳ nguyên âm nào.
6. Phủ Định Ký Tự [^ ]
Mô tả: Khớp với bất kỳ ký tự nào không nằm trong dấu ngoặc vuông.
Ví dụ: [^aeiou] sẽ khớp với bất kỳ ký tự nào không phải nguyên âm.
7. Phân Nhánh |
Mô tả: Khớp với một trong các biểu thức được ngăn cách bởi |.
Ví dụ: cat|dog sẽ khớp với "cat" hoặc "dog".
8. Ký Tự *, +, ?
Mô tả:
*: Khớp với phần trước đó 0 hoặc nhiều lần.
+: Khớp với phần trước đó 1 hoặc nhiều lần.
?: Khớp với phần trước đó 0 hoặc 1 lần.
Ví dụ:
ab* sẽ khớp với "a", "ab", "abb",...
ab+ sẽ khớp với "ab", "abb",...
ab? sẽ khớp với "a" hoặc "ab".
9. Cặp {m,n}
Mô tả: Khớp với phần trước đó ít nhất m lần và nhiều nhất n lần.
Ví dụ: ab{2,3} sẽ khớp với "abb" hoặc "abbb".
10. Cặp Ngoặc ( )
Mô tả: Nhóm các biểu thức để thực hiện các thao tác như kết hợp hoặc tham chiếu lại.
Ví dụ: (abc)+ sẽ khớp với "abc", "abcabc",...
11. Tham Chiếu Lại \num
Mô tả: Khớp với nội dung đã được nhóm lại trong cặp ngoặc ( ) lần thứ num.
Ví dụ: (ab)\1 sẽ khớp với "abab".
12. Các Ký Tự Đặc Biệt \d, \D, \w, \W, \s, \S
Mô tả:
\d: Khớp với bất kỳ chữ số nào (0-9).
\D: Khớp với bất kỳ ký tự nào không phải là chữ số.
\w: Khớp với bất kỳ ký tự chữ cái, chữ số, hoặc dấu gạch dưới.
\W: Khớp với bất kỳ ký tự nào không phải là chữ cái, chữ số, hoặc dấu gạch dưới.
\s: Khớp với bất kỳ ký tự khoảng trắng nào (space, tab, newline,...).
\S: Khớp với bất kỳ ký tự nào không phải là khoảng trắng.
Ví dụ:
\d+ sẽ khớp với "123", "4567",...
\w+ sẽ khớp với "word1", "var_name",...
13. Ký Tự \b và \B
Mô tả:
\b: Khớp với ranh giới từ.
\B: Khớp với nơi không phải ranh giới từ.
Ví dụ:
\bword\b sẽ khớp với từ "word" nhưng không phải là một phần của từ khác.
14. Chuỗi Thô r'string'
Mô tả: Chuỗi thô với tiền tố r giúp tránh phải thoát các ký tự đặc biệt.
Ví dụ: r'\d+' sẽ khớp với bất kỳ dãy chữ số nào mà không cần phải viết \\d+.
Những cú pháp trên là nền tảng của việc làm việc với biểu thức chính quy trong Python và giúp bạn xây dựng các mẫu tìm kiếm mạnh mẽ và linh hoạt.

```python
import re

# 1. Ký tự Thường
pattern = r'abc'
text = 'abc123'
match = re.search(pattern, text)
print(f"1. Ký tự Thường: {match.group()}")  # Kết quả: abc

# 2. Dấu Chấm `.`
pattern = r'a.c'
text = 'abc'
match = re.search(pattern, text)
print(f"2. Dấu Chấm `.`: {match.group()}")  # Kết quả: abc

# 3. Dấu Gạch Chéo `\`
pattern = r'\\d'  # Khớp với bất kỳ chữ số nào
text = '123abc'
match = re.search(pattern, text)
print(f"3. Dấu Gạch Chéo `\\`: {match.group()}")  # Kết quả: 1

# 4. Mốc Bắt Đầu `^` và Kết Thúc `$`
pattern_start = r'^hello'
text = 'hello world'
match_start = re.search(pattern_start, text)
print(f"4a. Mốc Bắt Đầu `^`: {match_start.group()}")  # Kết quả: hello

pattern_end = r'world$'
match_end = re.search(pattern_end, text)
print(f"4b. Kết Thúc `$`: {match_end.group()}")  # Kết quả: world

# 5. Ký Tự Trong Dấu Ngoặc Vuông `[ ]`
pattern = r'[aeiou]'
text = 'hello'
match = re.search(pattern, text)
print(f"5. Ký Tự Trong Dấu Ngoặc Vuông `[ ]`: {match.group()}")  # Kết quả: e

# 6. Phủ Định Ký Tự `[^ ]`
pattern = r'[^aeiou]'
text = 'hello'
matches = re.findall(pattern, text)
print(f"6. Phủ Định Ký Tự `[^ ]`: {matches}")  # Kết quả: ['h', 'l', 'l']

# 7. Phân Nhánh `|`
pattern = r'cat|dog'
text = 'I have a cat'
match = re.search(pattern, text)
print(f"7. Phân Nhánh `|`: {match.group()}")  # Kết quả: cat

# 8. Ký Tự `*`, `+`, `?`
pattern_star = r'ab*'
text = 'abbb'
match_star = re.search(pattern_star, text)
print(f"8a. Ký Tự `*`: {match_star.group()}")  # Kết quả: abbb

pattern_plus = r'ab+'
match_plus = re.search(pattern_plus, text)
print(f"8b. Ký Tự `+`: {match_plus.group()}")  # Kết quả: abbb

pattern_question = r'ab?'
match_question = re.search(pattern_question, text)
print(f"8c. Ký Tự `?`: {match_question.group()}")  # Kết quả: ab

# 9. Cặp `{m,n}`
pattern = r'ab{2,3}'
text = 'abbb'
match = re.search(pattern, text)
print(f"9. Cặp `{m,n}`: {match.group()}")  # Kết quả: abb

# 10. Cặp Ngoặc `( )`
pattern = r'(abc)+'
text = 'abcabcabc'
match = re.search(pattern, text)
print(f"10. Cặp Ngoặc `( )`: {match.group()}")  # Kết quả: abcabcabc

# 11. Tham Chiếu Lại `\\num`
pattern = r'(ab)\\1'
text = 'abab'
match = re.search(pattern, text)
print(f"11. Tham Chiếu Lại `\\num`: {match.group()}")  # Kết quả: abab

# 12. Các Ký Tự Đặc Biệt `\\d`, `\\D`, `\\w`, `\\W`, `\\s`, `\\S`
pattern_digit = r'\\d+'
text = 'There are 123 apples'
match_digit = re.search(pattern_digit, text)
print(f"12a. Ký Tự `\\d`: {match_digit.group()}")  # Kết quả: 123

pattern_word = r'\\w+'
match_word = re.search(pattern_word, text)
print(f"12b. Ký Tự `\\w`: {match_word.group()}")  # Kết quả: There
```
